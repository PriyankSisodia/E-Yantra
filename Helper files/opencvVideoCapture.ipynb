```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
capture = cv.VideoCapture(0)
while(capture.isOpened()):
    
    while True:
    ##capture frame by frame
        ret,frame = capture.read()   ##ret is either true or false (frame captured or not)
    
    ##operation on frame(convert into grayscale)
        grayFrame = cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    
    ##display the resulted frame
        cv.imshow('frame',grayFrame)
        cv.imshow('frame2',frame)
        if cv.waitKey(1) & 0xFF==ord('q'):
            break
##everthing is done ,release the capture
    capture.release()
    cv.destroyAllWindows()
else: 
    print("Alert ! Camera disconnected") 
```
